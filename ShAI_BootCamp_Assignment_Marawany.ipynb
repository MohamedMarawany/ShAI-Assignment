#Descriptive Statistics

```python
# Calculate basic statistics
df[['TotalPay']].describe()
```

```python
# Calculate mode
df['TotalPay'].mode()[0]
```

```python
# Determine the range of salaries
df['TotalPay'].max() - df['TotalPay'].min()
```

```python
# Calculate standard deviation
df['TotalPay'].std()
```

#Data Cleaning

```python
# Impute missing values in 'Benefits' column with mean
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

# Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Use a pie chart to represent the proportion of employees in different departments
department_distribution = df['JobTitle'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.pie(department_distribution, labels=department_distribution.index, autopct='%1.2f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# Grouped Analysis

```python
# Group the data by 'Year' and calculate average salaries for each year
df.groupby('Year')['TotalPay'].mean()
```

# Simple Correlation Analysis

```python
# Identify correlation between 'TotalPay' and 'BasePay'
correlation_salary_basepay = df['TotalPay'].corr(df['BasePay'])

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=df)
plt.title('Scatter Plot: Total Pay vs Base Pay')
plt.xlabel('Base Pay')
plt.ylabel('Total Pay')
plt.show()

print(f"\nCorrelation between TotalPay and BasePay: {correlation_salary_basepay}")
```
